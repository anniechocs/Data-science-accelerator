```python
import text_processing_demo
```

#### Load in the CPA data

```python
CPA = pd.read_csv("../data/processed/CPA_tokenized.csv")
CPA.drop('Unnamed: 0', axis=1, inplace=True)
# Create categories for Level 1 and Level 2 higherarchies
CPA2 = CPA.copy()
#get highest level of code
CPA2.loc[CPA2.Level !=1,'Category_2'] = CPA2[CPA2.Level !=1].Code.str.split('.').str.slice(0,1).str.join('')

# match up codes and parents
Code_parent = CPA2[CPA2.Level==2][['Parent','Category_2']].copy()
CPA2 = CPA2.merge(Code_parent.rename(columns={'Parent':'Category_1'}), on='Category_2', how='left')

# add in the Category_1 fileds for Level1
CPA2.loc[CPA2.Level==1, 'Category_1'] = CPA2.loc[CPA2.Level==1,'Code']

# we now set up a higher level for A10 indstry levels (10 categories)
update_dict0 = {'A':1,'F':3,'J':5, 'K':6, 'L':7,'M':8,'N':8}
update_dict = {**update_dict0,**dict.fromkeys(['B','C','D','E'],2),**dict.fromkeys(['G','H','I'],4),
               **dict.fromkeys(['O','P','Q'],9), **dict.fromkeys(['R','S','T','U'],10)}

CPA2['Category_0'] = CPA2.Category_1.replace(update_dict)
#CPA2 = CPA2.astype({'Category_0':int})
CPA_L6 = CPA2[CPA2.Level==6].copy()
CPA_L6.sample(5)
```

```python
df = text_processing_demo.clean_overview_col(CPA_L6, "tokens")
```

## Load FastText Pretrained

Note: This requires a fair bit of memory (peaks at about 17.5 GiB)

Recommend shutting down other kernels first, once this has loaded memory usage drops again.

This takes a few minutes to load in.

```python
wv = text_processing_demo.fetch_fasstext_pretrained(filepath="../../data/wiki.en.bin")
```

### Vectorize Movie overviews

```python
text_to_vec = functools.partial(text_processing_demo.vectorize_text, wv)

CPA_L6["tokens_cleaned_vectorized"] = CPA_L6.tokens_cleaned.apply(
    text_to_vec
)
```

### Dimensionality Reduction

```python
CPA_L6["tokens_cleaned_vectorized_low_dimension"] = text_processing_demo.reduce_dimensionality(
    CPA_L6.tokens_cleaned_vectorized
)
```

```python
CPA_L6.sample(5)
```

```python
plot_df = CPA_L6[["Category_2", "Code", "tokens_cleaned_vectorized_low_dimension"]].copy()

plot_df["x"] = plot_df.tokens_cleaned_vectorized_low_dimension.apply(lambda x: x[0])
plot_df["y"] = plot_df.tokens_cleaned_vectorized_low_dimension.apply(lambda x: x[1])

fig = plotly.express.scatter(
    plot_df, 
    x="x", 
    y="y", 
    color="Category_2",
    hover_data={
        "Code": True,
        "Category_2": False,
        "x": False,
        "y": False
        })
fig.update_layout(showlegend=False)
```

### Clustering

```python
CPA_L6["label"] = text_processing_demo.cluster(CPA_L6.tokens_cleaned_vectorized_low_dimension)
```

```python
plot_df = CPA_L6[["Category_0", "Code", "tokens_cleaned_vectorized_low_dimension","label"]].copy()

plot_df["x"] = plot_df.tokens_cleaned_vectorized_low_dimension.apply(lambda x: x[0])
plot_df["y"] = plot_df.tokens_cleaned_vectorized_low_dimension.apply(lambda x: x[1])

plot_df["label"] = plot_df.label.astype(str)

fig = plotly.express.scatter(
    plot_df, 
    x="x", 
    y="y", 
    color="label",
    hover_data={
        "Code": True,
        "label": True,
        "x": False,
        "y": False
        })
fig.update_layout(showlegend=False)
```
