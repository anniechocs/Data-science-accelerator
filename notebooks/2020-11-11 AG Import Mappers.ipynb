```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML
pd.set_option('display.max_colwidth', None)
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
data_path = '../data/processed/'
CN_CPA = pd.read_csv(data_path+'CN 2020-CPA 2.1.csv', skiprows=1)
CN_CPA['Source'] = pd.to_numeric(CN_CPA['Source'].str.replace(' ',''))

CPA_COI = pd.read_csv(data_path + 'COICOP 2018 - CPA 2 1.csv')
# read in CPA data
CPA = pd.read_csv(data_path + "CPA_data_cleaned.csv")
#CPA = CPA.drop(['Unnamed: 0','Also'],axis=1).rename(columns={'Includes_cleaned':'CPA_Includes','Excludes_cleaned':'CPA_Excludes'})
# Read in CN data file
CN = pd.read_csv(data_path+"CN_2020_full.csv", header=None, sep=',', names=['Code','Section','Description'])
CN['Code'] = np.floor_divide(CN['Code'],10000)
```

```python
Merge_CN_CPA = CN_CPA.rename(columns={'Source':'Code'}).merge(CN, on='Code')
Merge_CN_CPA.rename(columns={x : 'CN_'+ x for x in Merge_CN_CPA.columns}, inplace=True)
Merge_CN_CPA = Merge_CN_CPA.merge(CPA.rename(columns={'Code':'CN_Target'}), on='CN_Target')
Merge_CN_CPA.rename(columns={'CN_Target':'CPA_Code','Order':'CPA_Order','Descr':'CPA_Description'}, inplace=True)
#Merge_CN_CPA = Merge_CN_CPA[['CN_Code','CN_Section','CN_Description','CPA_Order','CPA_Code','CPA_Description','CPA_Includes','CPA_Excludes']]
Merge_CN_CPA.sample(6)
```

```python
Merge_CN_CPA.to_csv('CN_CPA_links.csv',index=False)
```

```python
CPA_COI = CPA_COI.replace('\\n',' ',regex=True)
CPA_COI = CPA_COI.replace('\t',' ',regex=True)
CPA = CPA.replace('\\n',' ',regex=True)
CPA = CPA.replace('\t',' ',regex=True)
CPA_COI = CPA_COI.replace('\r -',' ',regex=True)
CPA_COI.columns = ['Comments','COI_18','COI_18_1','COI_notes','COI_inc','COI_also_inc','COI_excl','CPA_Code','CPA_ex','CPA_Description','CPA_inc','CPA_also_inc','CPA_excl']
print(CPA_COI.columns)
CPA_COI.head()
CC = CPA_COI[['COI_18','COI_notes','COI_inc', 'COI_also_inc', 'COI_excl', 'CPA_Code', 'CPA_Description','CPA_inc', 'CPA_also_inc', 'CPA_excl']]
CC.sample(2)
```

```python
All_Merge = CC.merge(Merge_CN_CPA.drop(['CPA_Description','CPA_Includes','CPA_Excludes'],axis=1), on='CPA_Code')
All_Merge
```
